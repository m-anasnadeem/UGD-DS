```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income - canada_per_capita_income.csv")
df
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'])
plt.show()
```

```python
df2 = df.drop('per capita income (US$)', axis='columns')
df2
```

```python
price = df['per capita income (US$)']
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(df2,price)
```

```python
reg.predict([[2000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
year_df = pd.read_csv("years - years.csv")
year_df
```

```python
pred = reg.predict(year_df)
pred
```

```python
year_df['price']=pred
year_df
```

```python
prediction_df = pd.DataFrame({'year': year_df['year'], 'price': pred})

prediction_df.to_csv('prediction.csv', index=False)
```
